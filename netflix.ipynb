## PROBLEM STATEMENT
With the rapid expansion of digital streaming platforms,Netflix has developed a vast library of movies and TV shows across different countries,types and durations.However,understanding the patterns within this large datset is essential to identify trends in content production,audience targetting and platform growth.

This project aims to analyse the Netflix dataset to explore the distribution and characteristics of content available in the platform.The study focuses on identifying patterns in content type(Movies and TV shows),release trends over the years,country-wise contributions,ratings,classification and duration.


---

## ABOUT THE DATASET

- Data collected from kaggle
- The dataset used in this project is the Netflix Dataset. It contains information about movies and TV shows on Netflix.
- There are about 8,807 records in the dataset. It includes details like title, type, director, cast, country, release year, rating, duration, and genre.
- Some data was missing, so it was cleaned. The duration column was also corrected. Movies are measured in minutes, and TV shows are measured in seasons.
- This dataset helps us understand Netflix content, such as popular genres, release trends, and audience types. It is useful for learning data analysis.

---

## AIM OF THE ANALYSIS

The aim of this Project is to understand Netflix content. It helps to analyze movies and TV shows, find popular genres, study release year trends, and understand audience ratings. It is mainly used to learn data analysis and visualization.
